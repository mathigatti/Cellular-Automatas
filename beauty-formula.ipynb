```python
from fractaldimension import fractal_dimension
import cv2
import os
import number2automata

def fractalDimension(number):
    im = cv2.imread('images/'+str(number)+'.tiff', cv2.IMREAD_GRAYSCALE)
    newDimension = fractal_dimension(im, 0.9)
    return newDimension

def compressionScore(number):
    statinfo = os.stat('images/'+str(number)+'.png')
    png = statinfo.st_size
    
    statinfo = os.stat('images/'+str(number)+'.tiff')
    tiff = statinfo.st_size
    
    statinfo = os.stat('images/'+str(number)+'.jpg')
    jpg = statinfo.st_size
    
    return 100/png,100/tiff,100/jpg
```

```python
import random
numbers = list(range(1000000))
random.shuffle(numbers)
```

```python
scores = {}

for i in range(1000):
    number = numbers[i]
    number2automata.numberToAutomata(number,30)
    fscore = fractalDimension(number)
    cscore1, cscore2, cscore3 = compressionScore(number)
    scores[number] = {"Fractal Dimension":fscore,"Compression Eficciency 1":cscore1,"Compression Eficciency 2":cscore2,"Compression Eficciency 3":cscore3}
```

```python
import pandas as pd
df = pd.DataFrame(scores)
df = df.transpose()
df.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
clustering = AgglomerativeClustering(n_clusters=12).fit(df.values)
df["cluster"] = clustering.labels_
```

```python
df.plot(kind='scatter', x='Compression Eficciency 1', y='Compression Eficciency 2')
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

cmap = cm.get_cmap('gist_rainbow')

df.plot(kind='scatter', x='Fractal Dimension', y='Compression Eficciency 1',cmap=cmap, c='cluster')
```
